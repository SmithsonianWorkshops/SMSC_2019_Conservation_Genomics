## Plotting allele frequency spectra
Below is a method for plotting the allele (or site) frequency spectrum. We will plot two, one from ANGSD that used an ancestral genome (wild canary in this case) to estimate the 2D allele frequency spectrum and annother from GATK that didn't use an ancestral genome and so is folded.

First, we need to import dadi, which has some nice plotting tools for allele frequency spectra

```python
import dadi
```

Next, we have to read in the angsd allele frequency spectrum. This method is assuming that your allele frequency spectrum will only have the dimensions on the top row. In this case, 11 x 11 (N+1 on each axis).

```python
siskin_afs = dadi.Spectrum_mod.Spectrum.from_file("both_pop_angsd_dadi.sfs")
```

Then, we can use the plotting function to create a plot of the AFS

```python
dadi.Plotting.plot_single_2d_sfs(siskin_afs)
```

Now, let's do the same thing, but with the folded spectrum from GATK

```python
folded_afs_gatk = dadi.Spectrum_mod.Spectrum.from_file("Guy-Ven-gatk-4dadi.sfs")
```

```python
dadi.Plotting.plot_single_2d_sfs(folded_afs_gatk)
```
